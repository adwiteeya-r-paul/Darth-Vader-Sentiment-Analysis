```python
!pip install nltk
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
from google.colab import drive
drive.mount('/content/drive')

# dictionaries of emotions
d1= {}
d2 = {}
d3 = {}
d4 = {}
d5 = {}
d6 = {}
d7 = {}
d8 = {}
d9 = {}
d10 = {}


# tokenizers
vtokenizer = RegexpTokenizer(r'\w+')

# function to process scripts
# returns all darth vader lines as tokens

def process_script(file):
  vadertoken = []
  with open(file, 'r') as file:
    vader = "VADER"
    for line in file:
      if vader in line:
        line = line.split(vader)
        line = line[1].strip()
        line = vtokenizer.tokenize(line)
        vadertoken.append(line)
  return vadertoken


# function to split the emotion files into lists

def process_emotions(file):
  with open(file, 'r') as file:
    lists = []
    for line in file:
      line = line.strip()
      tokens = line.split()
      dict1(tokens)

#function that puts each token into the ten dictionaries
# with their scores as values

def dict1(tokens):
  if tokens == None:
    return 0
  if len(tokens) > 1:
    if tokens[1] == "anger":
      d1[tokens[0]] = tokens[2]
    if tokens[1] == "anticipation":
      d2[tokens[0]] = tokens[2]
    if tokens[1] == "disgust":
      d3[tokens[0]] = tokens[2]
    if tokens[1] == "fear":
      d4[tokens[0]] = tokens[2]
    if tokens[1] == "joy":
      d5[tokens[0]] = tokens[2]
    if tokens[1] == "negative":
      d6[tokens[0]] = tokens[2]
    if tokens[1] == "positive":
      d7[tokens[0]] = tokens[2]
    if tokens[1] == "sadness":
      d8[tokens[0]] = tokens[2]
    if tokens[1] == "surprise":
      d9[tokens[0]] = tokens[2]
    if tokens[1] == "trustdict":
      d10[tokens[0]] = tokens[2]


# function that calculates the total scores for each dictionary

def scores(tokens):

  angerscore = 0
  anticipationscore = 0
  disgustscore = 0
  fearscore = 0
  joyscore = 0
  negativescore = 0
  positivescore = 0
  sadnessscore = 0
  surprisescore = 0
  trustscore = 0

  for token in tokens:
    for word in token:
      if word in d1:
        angerscore += int(d1[word])
      if word in d2:
        anticipationscore += int(d2[word])
      if word in d3:
        disgustscore += int(d3[word])
      if word in d4:
        fearscore += int(d4[word])
      if word in d5:
        joyscore += int(d5[word])
      if word in d6:
        negativescore += int(d6[word])
      if word in d7:
        positivescore += int(d7[word])
      if word in d8:
        sadnessscore += int(d8[word])
      if word in d9:
        surprisescore += int(d9[word])
      if word in d10:
        trustscore += int(d10[word])

  print(("\n"))
  print("anger", angerscore)
  print("anticipation", anticipationscore)
  print("disgust", disgustscore)
  print("fear", fearscore)
  print("joy", joyscore)
  print("negative", negativescore)
  print("positive", positivescore)
  print("sadness", sadnessscore)
  print("surprise", surprisescore)
  print("trust", trustscore)
  print("TOTAL SCORE", angerscore + anticipationscore + disgustscore + fearscore + joyscore + negativescore + positivescore + sadnessscore + surprisescore + trustscore)
  print("\n")



# process emotions for the emotions file

process_emotions("/content/drive/My Drive/emotions.txt")

# tokenize all the scripts
tokens1 = process_script("/content/drive/My Drive/sw4.txt")
tokens2 = process_script("/content/drive/My Drive/sw5.txt")
tokens3 = process_script("/content/drive/My Drive/sw6.txt")


# print scores of all the scripts

print("\n")
print("-------sw4----------")
scores1 = scores(tokens1)
print("-------sw5----------")
scores2 = scores(tokens2)
print("-------sw6----------")
scores3 = scores(tokens3)
```
